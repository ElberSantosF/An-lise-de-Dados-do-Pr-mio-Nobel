```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('SDnobel.csv')
```

```python
df.select_dtypes(include = [int, float])
```

```python
df.sex.value_counts().plot(kind = "bar", rot = 0, edgecolor = "black", color = "tan")
plt.title("Ganhadores do prêmio nobel por gênero")
plt.xlabel("Gênero")
plt.ylabel("Quantidade")
plt.tight_layout()
```

```python
plt.figure(figsize = (8,6))
plt.scatter(df["age"], df["year"], alpha=0.6, color="gray")
plt.title("Prêmio Nobel: Relação entre idade e ano")
plt.xlabel("Idade")
plt.ylabel("Ano")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.age_group.value_counts().plot(kind = "bar", rot = 0, edgecolor = "black", color = "sandybrown")
plt.title("Ganhadores do Prêmio Nobel por Faixa Etária")
plt.xlabel("Faixa Etária")
plt.tight_layout()
```

```python
df.birth_country.value_counts().head(15).plot(kind="barh", ylabel="", title= "Prêmio Nobel: Top 15 países de nascimento", color = "darkblue")
plt.figure(figsize = (5,3))
plt.show()
```

```python
literature = df[df["category"] == 'Literature']
literature['birth_country'].value_counts().head(15).plot(kind='barh', ylabel="", color="cadetblue", edgecolor = "dimgrey")
plt.title('Top 15 Países Ganhadores do Nobel de Literatura')
plt.show()
```
